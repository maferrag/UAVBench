```python
# ============================
# 🚀 Developed by Dr. Mohamed Amine Ferrag @2025
# ============================
# Requirements: openai, pandas, matplotlib, tqdm, numpy

import os, json, re, time, random, queue, threading, string
from pathlib import Path
from datetime import datetime, timezone
from typing import Dict, Any, List, Optional, Tuple
import pandas as pd
import numpy as np
from google.colab import drive
from openai import OpenAI
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm
import matplotlib.pyplot as plt

# ---------- User Config ----------
DRIVE_JSON_DIR = "/content/drive/MyDrive/UAVBench_API/UAVBench_API_2000/scenario_mcqs"  # Folder containing MCQ JSONs

OPENROUTER_MODEL = os.getenv("OPENROUTER_MODEL", " ").strip()
OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY", "  ").strip()

visible_model_name = OPENROUTER_MODEL.replace("/", "／")  # full-width slash U+FF0F

RESULTS_CSV = f"/content/drive/MyDrive/UAVBench_API/UAVBench_API_2000/Evaluations/mcq_results_uavbench_{visible_model_name}.csv"
STATS_CSV   = f"/content/drive/MyDrive/UAVBench_API/UAVBench_API_2000/mcq_style_stats_uavbench_{visible_model_name}.csv"

# Concurrency & behavior
NUM_WORKERS = 20
BATCH_LIMIT = None   # e.g., 200 to limit
DRY_RUN     = False  # True = random answers
VERBOSE     = True   # print per-sample steps
PROGRESS_EVERY = 20

# ---------- Checks ----------
if not OPENROUTER_API_KEY and not DRY_RUN:
    raise RuntimeError("Missing OPENROUTER_API_KEY. Set via Colab secrets or os.environ.")

# ---------- Mount Drive ----------
drive.mount('/content/drive', force_remount=True)

# ---------- Client & helpers ----------
client = OpenAI(base_url="https://openrouter.ai/api/v1", api_key=OPENROUTER_API_KEY)
print("Model:", OPENROUTER_MODEL)

# Exactly 7 MCQ choices for UAVBench
LETTERS = list("ABCDEFG")   # A..G

def _utc_now_iso():
    return datetime.now(timezone.utc).isoformat(timespec="seconds").replace("+00:00", "Z")

def list_jsons(folder: str) -> List[Path]:
    p = Path(folder)
    return sorted([x for x in p.rglob("*.json") if x.is_file()])

def read_json(path: Path) -> Dict[str, Any]:
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

# ---------- Styles ----------
UAV_STYLES = {
    1: "Aerodynamics & Physics Reasoning",
    2: "Navigation & Path Planning Reasoning",
    3: "Mission Policy & Compliance Reasoning",
    4: "Environmental & Sensor Fusion Reasoning",
    5: "Multi-Agent Coordination Reasoning",
    6: "Cyber-Physical Security Reasoning",
    7: "Energy & Resource Management Reasoning",
    8: "Ethical & Safety-Critical Decision Reasoning",
    9: "Comparative System Reasoning",
    10:"Hybrid Integrated Reasoning",
}

# ---------- Normalization ----------
def strip_label_prefix(s: str) -> str:
    # remove leading "A- " / "B:" / "C " etc.
    return re.sub(r"^[A-Za-z][\s\-:]\s*", "", str(s).strip())

def normalize_schema(obj: Dict[str, Any], fallback_scenario_name: str) -> Dict[str, Any]:
    """
    Unified structure:
    {
      'schema_version','scenario_name','context','question',
      'choices' (list[str] length 7), 'num_choices' (7),
      'correct_letter' (str in A..G) or '',
      'style' (str), 'style_id' (int or None), 'gt_reason' (str)
    }
    """
    schema_version = str(obj.get("schema_version", "")).strip().lower()

    # UAVBench schema
    if schema_version.startswith("uavbench-mcq"):
        scenario_name = obj.get("scenario_name", fallback_scenario_name)
        context = obj.get("description", "")
        question = obj.get("question", "")
        raw_choices = obj.get("choices", []) or []
        choices = [strip_label_prefix(c) for c in raw_choices]
        # Enforce exactly 7 choices
        if len(choices) != 7:
            choices = (choices + [""]*7)[:7]
        num_choices = 7

        # Correct mapping: letter / index / text
        corr = obj.get("correct_choice", "")
        correct_letter = ""
        if isinstance(corr, str) and corr.strip():
            s = corr.strip()
            if len(s) == 1 and s.upper() in LETTERS:
                correct_letter = s.upper()
            else:
                t = strip_label_prefix(s)
                for idx, ch in enumerate(choices):
                    if t.strip().lower() == str(ch).strip().lower():
                        correct_letter = LETTERS[idx]
                        break
        elif isinstance(corr, int):
            if 0 <= corr < 7:
                correct_letter = LETTERS[corr]

        style_id = obj.get("style_id", None)
        style = obj.get("style", UAV_STYLES.get(style_id, "unknown"))
        gt_reason = obj.get("reason", "")

        return {
            "schema_version": obj.get("schema_version", "uavbench-mcq-v1"),
            "scenario_name": scenario_name,
            "context": context,
            "question": question,
            "choices": choices,
            "num_choices": num_choices,
            "correct_letter": correct_letter,
            "style": style or "unknown",
            "style_id": int(style_id) if isinstance(style_id, int) else None,
            "gt_reason": gt_reason,
        }

    # Fallback: original DriveBench-like schema (best-effort)
    desc = obj.get("description", "")
    question = obj.get("question", "")
    raw_choices = obj.get("choices", []) or []
    choices = [strip_label_prefix(c) for c in raw_choices]
    # Enforce 7 slots for uniform CSV (pad/trim)
    choices = (choices + [""]*7)[:7]
    correct_letter = ""
    corr = obj.get("correct_choice", "")
    if isinstance(corr, str):
        m = re.search(r"\b([A-G])\b", corr.strip(), flags=re.I)  # A..G only
        if m:
            correct_letter = m.group(1).upper()
        elif corr and corr.strip():
            t = strip_label_prefix(corr)
            for idx, ch in enumerate(choices):
                if t.strip().lower() == str(ch).strip().lower():
                    correct_letter = LETTERS[idx]
                    break
    elif isinstance(corr, int):
        if 0 <= corr < 7:
            correct_letter = LETTERS[corr]

    return {
        "schema_version": obj.get("schema_version", "drivebench-mcq"),
        "scenario_name": obj.get("scenario_name", fallback_scenario_name),
        "context": desc,
        "question": question,
        "choices": choices,
        "num_choices": 7,
        "correct_letter": correct_letter,
        "style": (obj.get("style") or "unknown"),
        "style_id": None,
        "gt_reason": "",
    }

# ---------- Prompting (A–G only) ----------
def build_messages(context: str, question: str, choices: List[str]) -> Tuple[List[Dict[str,str]], List[str]]:
    if len(choices) != 7:
        choices = (choices + [""]*7)[:7]
    labels = LETTERS  # A..G
    normalized = [f"{labels[i]}- {str(choices[i]).strip()}" for i in range(7)]
    choice_block = "\n".join(normalized)
    system = (
        "You are a careful evaluator for multiple-choice UAV mission questions. "
        "Return EXACTLY ONE CAPITAL LETTER from {A,B,C,D,E,F,G}. No words, punctuation, or explanation."
    )
    user = (
        (f"DESCRIPTION:\n{context}\n\n" if context else "") +
        f"QUESTION:\n{question}\n\n"
        f"CHOICES (pick ONE):\n{choice_block}\n\n"
        "Answer format: A single capital letter from {A,B,C,D,E,F,G} only."
    )
    return [{"role":"system","content":system},{"role":"user","content":user}], labels

def extract_allowed_letter(txt: str, allowed: List[str]) -> str:
    if not txt: return "?"
    s = txt.strip().upper()
    if s and s[0] in allowed:
        return s[0]
    for m in re.finditer(r"\b([A-G])\b", s):
        ch = m.group(1)
        if ch in allowed:
            return ch
    for ch in s:
        if ch in allowed:
            return ch
    return "?"

def call_openrouter_letter(messages: List[Dict[str, str]], model: str, allowed: List[str], max_retries=5, backoff_base=1.8) -> str:
    if DRY_RUN:
        return random.choice(allowed)
    for attempt in range(1, max_retries+1):
        try:
            resp = client.chat.completions.create(
                model=model,
                messages=messages,
                temperature=0.0,
                max_tokens=16,  # single letter + margin
                top_p=1.0,
            )
            txt = (resp.choices[0].message.content or "").strip()
            letter = extract_allowed_letter(txt, allowed)
            return letter
        except Exception as e:
            delay = min(30, (backoff_base ** attempt)) + random.uniform(0, 0.75)
            if VERBOSE:
                print(f"[retry] {e} — sleeping {delay:.1f}s (attempt {attempt})")
            time.sleep(delay)
    return "?"

# ---------- CSV Helpers ----------
def ensure_csv(path: str):
    p = Path(path)
    if not p.exists():
        df = pd.DataFrame(columns=[
            "timestamp","file","schema_version","scenario_name","model",
            "style_id","style","num_choices",
            "answer","correct_letter","is_correct",
            "question","context",
            "choice_A","choice_B","choice_C","choice_D","choice_E","choice_F","choice_G",
            "choices_json","gt_reason"
        ])
        df.to_csv(p, index=False)

def already_done_set(path: str) -> set:
    p = Path(path)
    if not p.exists():
        return set()
    try:
        df = pd.read_csv(p, usecols=["file"])
        return set(df["file"].astype(str).tolist())
    except Exception:
        return set()

# ---------- Load file list ----------
ensure_csv(RESULTS_CSV)
done = already_done_set(RESULTS_CSV)
json_files = list_jsons(DRIVE_JSON_DIR)
if BATCH_LIMIT is not None:
    json_files = json_files[:BATCH_LIMIT]

print(f"Found {len(json_files)} JSON files in {DRIVE_JSON_DIR}")
print(f"Already in CSV: {len(done)}")

# ---------- Writer thread ----------
write_q: "queue.Queue[Dict[str, Any]]" = queue.Queue(maxsize=1000)
stop_signal = object()
csv_lock = threading.Lock()

def writer_loop():
    buffer = []
    last_flush = time.time()
    FLUSH_EVERY_SEC = 3
    while True:
        item = write_q.get()
        if item is stop_signal:
            break
        buffer.append(item)
        if len(buffer) >= 64 or (time.time() - last_flush) >= FLUSH_EVERY_SEC:
            dfw = pd.DataFrame(buffer)
            with csv_lock:
                header_needed = (not Path(RESULTS_CSV).exists()) or os.path.getsize(RESULTS_CSV) == 0
                dfw.to_csv(RESULTS_CSV, mode="a", header=header_needed, index=False)
            buffer.clear()
            last_flush = time.time()
    if buffer:
        dfw = pd.DataFrame(buffer)
        with csv_lock:
            header_needed = (not Path(RESULTS_CSV).exists()) or os.path.getsize(RESULTS_CSV) == 0
            dfw.to_csv(RESULTS_CSV, mode="a", header=header_needed, index=False)

writer_thread = threading.Thread(target=writer_loop, daemon=True)
writer_thread.start()

# ---------- Worker ----------
def process_file(path: Path) -> Dict[str, Any]:
    rel = str(path)
    try:
        obj = read_json(path)
        if VERBOSE: print(f"[read] {rel}")
    except Exception as e:
        return {"error": f"read_json failed: {e}", "file": rel}

    norm = normalize_schema(obj, fallback_scenario_name=path.stem)

    # validation
    if not (norm["question"] and isinstance(norm["choices"], list) and len(norm["choices"]) >= 2):
        return {"error": "missing fields", "file": rel}

    if VERBOSE:
        print(f"[prompt] {rel} | style={norm['style']} | scenario={norm['scenario_name']} | N={norm['num_choices']}")

    messages, allowed_labels = build_messages(norm["context"], norm["question"], norm["choices"])

    answer = call_openrouter_letter(messages, OPENROUTER_MODEL, allowed_labels)
    if VERBOSE:
        print(f"[llm] {rel} → {answer}")

    correct = (norm["correct_letter"] or "").upper()
    is_correct = (answer == correct) if correct else ""

    padded = (norm["choices"] + [""]*7)[:7]
    row = {
        "timestamp": _utc_now_iso(),
        "file": rel,
        "schema_version": norm["schema_version"],
        "scenario_name": norm["scenario_name"],
        "model": OPENROUTER_MODEL,
        "style_id": norm["style_id"],
        "style": (norm["style"] or "unknown").strip(),
        "num_choices": 7,
        "answer": answer,
        "correct_letter": correct,
        "is_correct": is_correct,
        "question": norm["question"],
        "context": norm["context"],
        "choice_A": padded[0],
        "choice_B": padded[1],
        "choice_C": padded[2],
        "choice_D": padded[3],
        "choice_E": padded[4],
        "choice_F": padded[5],
        "choice_G": padded[6],
        "choices_json": json.dumps(norm["choices"], ensure_ascii=False),
        "gt_reason": norm["gt_reason"],
    }
    if VERBOSE:
        print(f"[write] {rel} | is_correct={is_correct}")
    return row

# ---------- Parallel execution ----------
to_process = [p for p in json_files if str(p) not in done]
print(f"Queued {len(to_process)} files for processing with {NUM_WORKERS} threads.")

processed = 0
errors = 0

with ThreadPoolExecutor(max_workers=NUM_WORKERS) as ex:
    futures = {ex.submit(process_file, p): p for p in to_process}
    for fut in tqdm(as_completed(futures), total=len(futures), desc="Processing"):
        res = fut.result()
        if "error" in res:
            errors += 1
            if VERBOSE:
                print(f"[error] {res.get('file','?')}: {res['error']}")
        else:
            write_q.put(res)
            processed += 1
            if processed % PROGRESS_EVERY == 0:
                print(f"[progress] processed={processed}/{len(to_process)} | errors={errors}")

write_q.put(stop_signal)
writer_thread.join()

print(f"✅ Done. Processed={processed}, Errors={errors}")
print(f"Results CSV → {RESULTS_CSV}")

# ---------- Accuracy by UAVBench style ----------
if not Path(RESULTS_CSV).exists():
    raise FileNotFoundError(f"Results CSV not found: {RESULTS_CSV}")

df = pd.read_csv(RESULTS_CSV)

for col in ["style","is_correct"]:
    if col not in df.columns:
        raise ValueError(f"Missing column '{col}' in {RESULTS_CSV}")

def norm_style(s):
    if pd.isna(s): return "unknown"
    ss = str(s).strip()
    if ss in UAV_STYLES.values(): return ss
    return "unknown"

df["style"] = df["style"].apply(norm_style)

def to_bool(x):
    if pd.isna(x): return np.nan
    s = str(x).strip().lower()
    if s in ("true","1","yes","y"):  return True
    if s in ("false","0","no","n"):  return False
    if x in (True, False):           return bool(x)
    return np.nan

df["is_correct_bool"] = df["is_correct"].apply(to_bool)

agg = (
    df.groupby("style", dropna=False)
      .agg(
          total   = ("is_correct_bool", "size"),
          answered= ("is_correct_bool", lambda s: int(s.notna().sum())),
          correct = ("is_correct_bool", lambda s: int((s == True).sum())),
          incorrect=("is_correct_bool", lambda s: int((s == False).sum())),
      )
      .reset_index()
)

agg["accuracy"] = (agg["correct"] / agg["answered"]).replace([np.inf, -np.inf], np.nan).round(4)

overall = pd.DataFrame([{
    "style": "ALL",
    "total": int(df.shape[0]),
    "answered": int(df["is_correct_bool"].notna().sum()),
    "correct": int((df["is_correct_bool"] == True).sum()),
    "incorrect": int((df["is_correct_bool"] == False).sum()),
}])
overall["accuracy"] = (overall["correct"] / overall["answered"]).round(4)

summary = pd.concat([agg, overall], ignore_index=True)
summary.to_csv(STATS_CSV, index=False)

print(summary)
print("\n📁 Saved summary to:", STATS_CSV)

# ---------- Optional chart: the 10 UAVBench styles ----------
view = summary[summary["style"].isin(UAV_STYLES.values())]
if not view.empty:
    plt.figure()
    plt.bar(view["style"], view["accuracy"])
    plt.title("UAVBench — Accuracy by Reasoning Style")
    plt.xlabel("Reasoning Style")
    plt.ylabel("Accuracy")
    plt.xticks(rotation=60, ha="right")
    plt.tight_layout()
    plt.show()
```
